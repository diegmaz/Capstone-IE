```python
import pandas as pd
import geopandas as gpd
from shapely.geometry import Point

# Load the new dataset
df = pd.read_csv('olist_final.csv')

state_name_mapping = {
    'Acre': 'Acre',
    'Alagoas': 'Alagoas',
    'Amapa': 'Amapá',
    'Amazonas': 'Amazonas',
    'Bahia': 'Bahia',
    'Ceara': 'Ceará',
    'Distrito Federal': 'Distrito Federal',
    'Espirito Santo': 'Espírito Santo',
    'Goias': 'Goiás',
    'Maranhao': 'Maranhão',
    'Mato Grosso': 'Mato Grosso',
    'Mato Grosso do Sul': 'Mato Grosso do Sul',
    'Minas Gerais': 'Minas Gerais',
    'Para': 'Pará',
    'Paraiba': 'Paraíba',
    'Parana': 'Paraná',
    'Pernambuco': 'Pernambuco',
    'Piaui': 'Piauí',
    'Rio de Janeiro': 'Rio de Janeiro',
    'Rio Grande do Norte': 'Rio Grande do Norte',
    'Rio Grande do Sul': 'Rio Grande do Sul',
    'Rondonia': 'Rondônia',
    'Roraima': 'Roraima',
    'Santa Catarina': 'Santa Catarina',
    'Sao Paulo': 'São Paulo',
    'Sergipe': 'Sergipe',
    'Tocantins': 'Tocantins'
}

# Replace state names in the dataframe with the correct names
df['geolocation_state'] = df['geolocation_state'].replace(state_name_mapping)
# Convert to GeoDataFrame
geometry = [Point(xy) for xy in zip(df['geolocation_lng'], df['geolocation_lat'])]
gdf = gpd.GeoDataFrame(df, geometry=geometry)

# Load the GeoJSON file containing state boundaries
gdf_states = gpd.read_file('brazil-states.geojson')

# Ensure the CRS (Coordinate Reference System) matches
gdf.crs = gdf_states.crs

# Inspect the columns to find the state name columns
print("Geolocation DataFrame (gdf):")
print(gdf.head())
print(gdf.columns)
print("\nGeoJSON States DataFrame (gdf_states):")
print(gdf_states.head())
print(gdf_states.columns)

# Remove existing index columns if any
gdf = gdf.drop(columns=gdf.filter(regex='index').columns, errors='ignore')
gdf_states = gdf_states.drop(columns=gdf_states.filter(regex='index').columns, errors='ignore')

# Perform the spatial join
filtered_gdf = gpd.sjoin(gdf, gdf_states, how='inner', predicate='within')

# Display the columns of the joined DataFrame to ensure the state name column is present
print("\nJoined DataFrame (filtered_gdf):")
print(filtered_gdf.head())
print(filtered_gdf.columns)

# Ensure the 'name' column is in the joined dataframe
if 'name' not in filtered_gdf.columns:
    raise KeyError("Error: 'name' column not found in the joined GeoDataFrame")

# Remove points with mismatched states (assuming 'geolocation_state' matches 'name')
filtered_gdf = filtered_gdf[filtered_gdf['geolocation_state'] == filtered_gdf['name']]

# Drop unnecessary columns from the join
filtered_gdf = filtered_gdf.drop(columns=['index_right', 'name', 'regiao_id', 'codigo_ibg', 'cartodb_id', 'created_at', 'updated_at'], errors='ignore')

# Save the cleaned dataset
filtered_gdf.to_csv('permanently_cleaned_geolocation_data.csv', index=False)

print("Dataset cleaned and saved as 'permanently_cleaned_geolocation_data.csv'")
```

```python
import pandas as pd
import dash
from dash import dcc, html
from dash.dependencies import Input, Output
import plotly.express as px
from sklearn.cluster import KMeans

# Load the cleaned dataset
df = pd.read_csv('permanently_cleaned_geolocation_data.csv')

# Initialize the Dash app
app = dash.Dash(__name__, external_stylesheets=['https://codepen.io/chriddyp/pen/bWLwgP.css'])

# State Filter Options with "Select All"
state_options = [{'label': 'Select All', 'value': 'all'}] + [
    {'label': i, 'value': i} for i in df['geolocation_state'].unique()
]

app.layout = html.Div([
    html.H1("Olist Warehouse Location Optimization Tool", style={'textAlign': 'center'}),

    # Outer container for slicers
    html.Div(id='outer_div', children=[
        html.Div([  # Slicer 1 (Object)
            html.Label("Select Object:"),
            dcc.Dropdown(id='object-filter', options=[{'label': i, 'value': i} for i in df['object'].unique()], multi=True)
        ], style={'width': '250px', 'margin': '10px'}),  # Adjust width as needed

        html.Div([  # Slicer 2 (State)
            html.Label("Select State:"),
            dcc.Dropdown(id='state-filter', options=state_options, value='all', multi=True)
        ], style={'width': '250px', 'margin': '10px'}),

        html.Div([  # Slicer 3 (Number of Clusters)
            html.Label("Number of Clusters:"),
            dcc.Input(id='cluster-input', type='number', min=1, value=3)
        ], style={'width': '250px', 'margin': '10px'}),
    ], style={
        'display': 'flex',       
        'flex-direction': 'row',  
        'justify-content': 'center',
        'align-items': 'center',
        'flex-wrap': 'wrap'
    }),

    html.Div(  # Map container
        dcc.Graph(id='cluster-plot', style={'height': '600px', 'width': '800px'}), 
        style={
            'display': 'flex',
            'justify-content': 'center', 
            'align-items': 'center'  
        }
    )
])

# Callback for Interactivity
@app.callback(
    Output('cluster-plot', 'figure'),
    [Input('object-filter', 'value'),
     Input('state-filter', 'value'),
     Input('cluster-input', 'value')]
)
def update_plot(selected_objects, selected_states, num_clusters):
    # Handle "Select All" for States
    if 'all' in selected_states:
        selected_states = df['geolocation_state'].unique()

    if selected_objects is None or num_clusters is None:
        return px.scatter_mapbox()  # Empty plot if filters are not selected

    filtered_df = df[df['object'].isin(selected_objects) & df['geolocation_state'].isin(selected_states)]

    if filtered_df.empty:
        return px.scatter_mapbox()

    # Clustering
    kmeans = KMeans(n_clusters=num_clusters).fit(filtered_df[['geolocation_lng', 'geolocation_lat']])
    filtered_df['cluster'] = kmeans.labels_

    # Calculate Cluster Centers
    cluster_centers = filtered_df.groupby('cluster')[['geolocation_lng', 'geolocation_lat']].mean().reset_index()

    # Create Plot (NO hovertemplate here)
    fig = px.scatter_mapbox(filtered_df, lat="geolocation_lat", lon="geolocation_lng", color="cluster",
                            hover_name=None,
                            custom_data=['cluster'],
                            mapbox_style="carto-positron")

    # Add Cluster Centers to Plot (WITH hover data)
    fig.add_scattermapbox(
        lat=cluster_centers["geolocation_lat"],
        lon=cluster_centers["geolocation_lng"],
        mode="markers",
        marker=dict(size=20, color="red"),
        name="Warehouses",
        hoverinfo="text",  # Enable hover for cluster centers
        text=[
            f"Warehouse {i + 1}<br>Lat: {lat:.4f}<br>Lon: {lon:.4f}"
            for i, lat, lon in zip(
                cluster_centers.index, cluster_centers["geolocation_lat"], cluster_centers["geolocation_lng"]
            )
        ],
        showlegend=False,
    )

    return fig

if __name__ == '__main__':
    app.run_server(debug=True, port=8055)
```
